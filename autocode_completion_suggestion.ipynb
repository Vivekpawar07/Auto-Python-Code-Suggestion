```python
import tensorflow as tf
import tensorflow.keras.preprocessing.text as Tokenizer
```

```python
with open('/content/input_9.txt','r') as f:
  data = f.read()
print(data)
```

```python
def cleanData(data):
  df = []
  curr_df = []
  for i in data.splitlines():
    if i.strip() == "":
      if curr_df:
        df.append("\n".join(curr_df))
        curr_df = []

    else:
      curr_df.append(i)
  if curr_df:
        df.append("\n".join(curr_df))
  df = [x + ' \n' for x in df]
  x = ""
  df = x.join(df)
  return df
```

```python
x = cleanData(data)
print(x)
```

```python
tokenizer = Tokenizer.Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
input_sentence = []
for i in data.split("/n"):
  tokenized = tokenizer.texts_to_sequences([i])[0]

  for i in range(1,len(tokenized)):
    input_sentence.append(tokenized[:i+1])
```

```python
print(input_sentence)
```

```python
max_len = max([len(i) for i in input_sentence])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequences = pad_sequences(input_sentence, maxlen=max_len, padding='pre')
padded_sequences
```

```python
x = padded_sequences[:,:-1]
y = padded_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
unique = len(tokenizer.word_index)
unique
```

```python
from tensorflow.keras.utils import to_categorical
y_encoded = tf.keras.utils.to_categorical(y, num_classes=958)
y_encoded
```

```python
y_encoded.shape
```

```python
from keras import Sequential
from keras.layers import Dense, LSTM, Embedding,BatchNormalization
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Embedding(input_dim=957, output_dim=100))
model.add(LSTM(128, return_sequences=True))
model.add(BatchNormalization())
model.add(LSTM(64))
model.add(Dense(32, activation='relu'))
model.add(Dense(958, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
callback = EarlyStopping(monitor='loss', patience=3)
```

```python
model.fit(x,y_encoded, epochs=100, callbacks=[callback])
```

```python
model.save('code_completion')
```
